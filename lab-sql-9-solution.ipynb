#### 1. How many distinct (different) actors' last names are there?

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/bank'
engine = create_engine(connection_string)
data = pd.read_sql_query('select count(distinct(last_name)) as different_names from sakila.actor;', engine)
data.head()
```

#### 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
query2 = '''select *, weekday(rental_date) as 'Weekday',
case
when (weekday(rental_date) >  4 ) then 'Weekend'
else 'Workday'
end as day_type
from sakila.rental
order by rental_date desc;'''
data = pd.read_sql_query(query2, engine)
data.head()
```

#### 3. Get all films with ARMAGEDDON in the title.

```python
data = pd.read_sql_query('''select *  from sakila.film where title like '%%ARMAGEDDON%%';''', engine)
data.head()
```

#### 4. Get 10 the longest films.

```python
query4 = 'select * from sakila.film order by length desc limit 10;'
data = pd.read_sql_query(query4, engine)
data.head()
```

#### 5. How many films include Behind the Scenes content?

```python
query5 = '''select film_id, title, special_features from sakila.film where special_features like '%%Behind_the_Scenes%%';'''
data = pd.read_sql_query(query5, engine)
data.head()
```

#### 6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
query6 = 'select rating, avg(length) from sakila.film group by rating having avg(length) > 120 ;'
data = pd.read_sql_query(query6, engine)
data.head()
```

#### 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query7 = '''select title as 'movie title', length, RANK() OVER(order by length desc) as `movie rank` FROM sakila.film;'''
data = pd.read_sql_query(query7, engine)
data.head()
```
